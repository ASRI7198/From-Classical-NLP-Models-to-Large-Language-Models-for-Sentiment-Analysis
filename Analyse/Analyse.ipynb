```python
train_df = fn.load_data(args.PATH_DATA_TRAIN)
val_df   = fn.load_data(args.PATH_DATA_VAL)
test_df  = fn.load_data(args.PATH_DATA_TEST)
```

```python
print("Number of classes for train dataset:", train_df['label'].nunique())
print("Number of classes for validation dataset:", val_df['label'].nunique())
print("Number of classes for test dataset:", test_df['label'].nunique())
```

```python
label_counts_train = train_df['label'].value_counts()
print(label_counts_train)
```

```python
label_counts_val = val_df['label'].value_counts()
print(label_counts_val)
```

```python
label_counts_test = test_df['label'].value_counts()
print(label_counts_test)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=label_counts_train.index, y=label_counts_train.values, palette='viridis')
plt.title('Distribution of Classes in Training Dataset')
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=label_counts_val.index, y=label_counts_val.values, palette='viridis')
plt.title('Distribution of Classes in Validation Dataset')
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=label_counts_test.index, y=label_counts_test.values, palette='viridis')
plt.title('Distribution of Classes in Test Dataset')
```
